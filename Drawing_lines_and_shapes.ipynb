```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a canvas
height,width = 500,700
blue=(255,127,0)
canvas = np.full((height,width,3),blue,dtype = np.uint8)

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
#drawing a LIne
cv2.line(canvas, (50,50),(650,450),(255,255,255),1)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
# drawing a rectangle
cv2.rectangle(canvas,(100,100),(300,300), (0,255,255),-1)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
#drawing a circle
cv2.circle(canvas, (500,150), 100 ,(0,0,255),5)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
cv2.ellipse(canvas, (350,350), (100,50) , 45 ,100 ,360, (0,255,0) ,3 )
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
points = np.array([[150,400],[300,400],[255,300]], np.int32)
new_reshaped_point = points.reshape((-1,1,2))
print(new_reshaped_point)
```

```python
cv2.polylines(canvas, [new_reshaped_point],True,(255,0,255),3)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```
